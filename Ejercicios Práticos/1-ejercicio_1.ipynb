### EJERCICIO 1:

#### SOLUCIN DEL PROBLEMA A:

- DIferencia en porcentaje del ms rpido de otros cursos

```python
horas_este_curso = 1.5 # Datos en horas
horas_mas_rapido_otros_curso = 2.5 # Dato en horas
    
# DIFERENCIA PORCENTUAL = ((VALOR FINAL - VALOR INICIAL) / VALOR INCIAL) * 100
diferencia_porcentaje = ((horas_mas_rapido_otros_curso - horas_este_curso)) / horas_este_curso * 100
print(f"La diferencia en porcentaje entre el curso actual y el ms rpido de otros cursos es de: {diferencia_porcentaje:.2f} %.")
```

- Diferencia del ms lento de otros cursos

```python
horas_mas_lento_otros_curso =  7 # Dato en horas

# DIFERENCIA PORCENTUAL = ((VALOR FINAL - VALOR INICIAL) / VALOR INCIAL) * 100
diferencia_porcentaje = ((horas_mas_lento_otros_curso - horas_este_curso)) / horas_este_curso * 100
print(f"La diferencia en porcentaje entre el curso actual y el ms lento de otros cursos es de: {diferencia_porcentaje:.2f} %")
```

- Diferencia del promedio de otros cursos

```python
horas_promedio_otros_curso =  4 # Dato en horas

# DIFERENCIA PORCENTUAL = ((VALOR FINAL - VALOR INICIAL) / VALOR INCIAL) * 100
diferencia_porcentaje = ((horas_promedio_otros_curso - horas_este_curso) / horas_este_curso) * 100
print(f"La diferencia en porcentaje entre el curso actual y el promedio de otros cursos es de: {diferencia_porcentaje:.2f} %")
```

#### SOLUCIN DEL PROBLEMA B:

- El promedio de los cursos se reduce en:

```python
promedio_crudo_otro_cursos =  5 # Dato en horas

# DIFERENCIA PORCENTUAL = ((VALOR MAYOR - VALOR MENOR) / VALOR MAYOR) * 100
diferencia_porcentaje = ((promedio_crudo_otro_cursos - horas_promedio_otros_curso)) / promedio_crudo_otro_cursos * 100
print(f"El porcentaje de material inservible del promedio de otros cursos: {diferencia_porcentaje:.2f} %")
```

- El promedio de este curso se reduce en:

```python
promedio_crudo_este_curso =  3.5 # Dato en horas

# DIFERENCIA PORCENTUAL = ((VALOR MAYOR - VALOR MENOR) / VALOR MAYOR) * 100
diferencia_porcentaje = ((promedio_crudo_este_curso - horas_este_curso)) / promedio_crudo_este_curso * 100
print(f"El porcentaje de material inservible del promedio de este curso: {diferencia_porcentaje:.2f} %")
```
