### EJERCICIO 2:

#### SOLUCIÓN DEL PROBLEMA A:

- Ingresar un texto y calcular cuanto tardaría en decir la frase

```python
promedio_palabras_por_segundo = 2
texto = input("Ingresar texto: ")
palabras = texto.split()
cantidad_palabras = len(palabras)

tiempo_en_decir_frase = cantidad_palabras / promedio_palabras_por_segundo


print(f"El tiempo de una persona normal en decir el texto '{texto}' es de: {tiempo_en_decir_frase} segundos.")
print(f"Y dijo un total de: {cantidad_palabras} palabra(s)")
```

#### SOLUCIÓN DEL PROBLEMA B:

- Si tarfa mas de 1 minuto decirle: "para flaco tampoco te pedí un testamento"

```python
promedio_palabras_por_segundo = 2
texto = input("Ingresar texto: ")
palabras = texto.split()
cantidad_palabras = len(palabras)

tiempo_en_decir_frase = cantidad_palabras / promedio_palabras_por_segundo

if tiempo_en_decir_frase > 60:
    print("Para flaco tampoco te pedí un testamento.")
```

#### SOLUCIÓN DEL PROBLEMA C:

- Si Dalto habla un 30% más rápido, ¿Cuánto tardaría en decirlo?

```python
promedio_palabras_por_segundo = 2
texto = input("Ingresar texto: ")
palabras = texto.split()
cantidad_palabras = len(palabras)

tiempo_en_decir_frase = cantidad_palabras / promedio_palabras_por_segundo


print(f"El tiempo de una persona normal en decir el texto '{texto}' es de: {tiempo_en_decir_frase} segundos.")
print(f"Y dijo un total de: {cantidad_palabras} palabra(s)")
print(f"Dalto tardaría {tiempo_en_decir_frase - tiempo_en_decir_frase*.3}")
```
